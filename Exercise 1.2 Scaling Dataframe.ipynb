## Change Date and Month Types

```python
df.rename(columns={'DATE':'intDATE'}, inplace=True)
```

```python
df.head()
```

```python
df['DATE'] = pd.to_datetime(df['intDATE'].astype(str), format='%Y%m%d')
```

```python
df.head()
```

```python
df = df.drop('intDATE', axis=1)
```

```python
df.head()
```

```python
df['MONTH'] = df['MONTH'].astype('str')
```

```python
df.head()
```

## StandardScaler

```python
scaler = StandardScaler()
```

```python
cols = df.columns[df.columns != 'DATE']
```

```python
#Create a new df with scaled data
df_stdscaled = pd.DataFrame(scaler.fit_transform(df[cols]))
df_stdscaled.head()
```

```python
# Compare the original vs. scaled data

df.head()
```

```python
# Export dataframe as a csv file
df_scaled.to_csv(os.path.join(path, '02 Data', 'Prepared Data', 'Dataset_stdscaled.csv'))
```

```python
scaler = MinMaxScaler()
```
