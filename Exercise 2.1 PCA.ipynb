## Dendrograms

###   Single Method - Unscaled

```python


distance_single = linkage(dfyear.loc[:,
['PCA1',
 'PCA2',
 'PCA3',
 'PCA4',
 'PCA5',
 'PCA6',
 'PCA7',
 'PCA8',
 'PCA9']],method="single")
plt.figure(figsize=(30,10))
dendrogram(distance_single, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Dendrogram Single Method - Unscaled",fontsize=18)
plt.show()
```

```python
# 12 clusters found
```
